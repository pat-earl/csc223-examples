```python
%matplotlib inline
import numpy as np
import matplotlib.pyplot as plt

plt.style.use('seaborn-ticks')
```

```python
plt.style.available
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
fig.savefig('figure.png')
```

```python
# fig2 = plt.figure()
plt.plot(x, np.tan(x))
```

```python
plt.figure()
plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x), color='green')
ax[1].plot(x, np.cos(x), color="#21dffc")
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x), color='green', linestyle='dashdot')
ax[1].plot(x, np.cos(x), color="#21dffc", linestyle=':')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), 'o', label='sin wave')
plt.plot(x, np.cos(x), label='cos wave')

plt.xlim(-2, 12)
plt.ylim(-1.25, 1.25)

plt.title("First Plot")
plt.xlabel("x")
plt.ylabel("y")
plt.legend()
```

```python
y = np.sin(x)
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=0.5)
plt.ylim(-1.2, 1.2)
```

```python
rng = np.random.RandomState(0)
x = rng.randn(100)
y = rng.randn(100)
colors = rng.rand(100)
sizes = 1000 * rng.rand(100)

plt.scatter(x, y, c=colors, s=sizes, alpha=0.3,
            cmap='viridis')
plt.colorbar()
```
