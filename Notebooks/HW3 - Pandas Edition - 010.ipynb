```python
import pandas as pd
```

```python
df = pd.read_csv('netflix_titles.csv', index_col=0)
df.head()
```

```python
tv_shows = df[df["type"] == "TV Show"].copy()
tv_shows.head()
```

```python
tv_shows = tv_shows[['title', 'director', 'country', 'date_added', 'duration']]
```

```python
tv_shows.head()
```

```python
tv_shows['duration'] = tv_shows['duration'].str.split(' ').str[0]
tv_shows['duration'] = tv_shows['duration'].astype('int32')
tv_shows['duration']
```

```python
tv_shows = tv_shows.sort_values('duration', ascending=False)
```

```python
tv_shows.to_json('netflix_titles.json')
```
