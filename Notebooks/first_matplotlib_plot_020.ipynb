```python
%matplotlib inline

import numpy as np
import matplotlib.pyplot as plt
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x))
plt.plot(x, np.cos(x))
```

```python
fig.savefig('figure.png')
```

```python
plt.figure()

plt.subplot(2, 1, 1)
plt.plot(x, np.sin(x))

plt.subplot(2, 1, 2)
plt.plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x))
ax[1].plot(x, np.cos(x))
```

```python
fig, ax = plt.subplots(2)
ax[0].plot(x, np.sin(x), color='green', linestyle='dashed')
ax[1].plot(x, np.cos(x), color="#20f000", linestyle=':')
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), label="Sin Wave", color='green')
plt.plot(x, np.cos(x), label="Cos Wave")

plt.xlim(-2, 12)
plt.ylim(-1.25, 1.5)

plt.title("My First Plot")
plt.xlabel("X")
plt.ylabel("Y")

plt.legend()
```

```python
x = np.linspace(0, 10, 100)

fig = plt.figure()
plt.plot(x, np.sin(x), 'o')
```

```python
y = np.sin(x)
plt.plot(x, y, '-p', color='gray',
         markersize=15, linewidth=4,
         markerfacecolor='white',
         markeredgecolor='gray',
         markeredgewidth=2)
plt.ylim(-1.2, 1.2)
```
