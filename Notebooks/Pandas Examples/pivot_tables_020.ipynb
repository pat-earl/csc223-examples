```python
import pandas as pd
import seaborn as sns
titanic = sns.load_dataset('titanic')
```

```python
cars = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/DAAG/toycars.csv', index_col=0)
cars.head()
```

```python
titanic.head()
```

```python
titanic.groupby(['sex', 'class'])[['survived']].mean().unstack()
```

```python
titanic.pivot_table('survived', index='sex', columns='class')
```

```python
cars.head()
```

```python
new_df = cars.pivot(values='distance', index='angle', columns='car')
```

```python
stats = cars.groupby('angle')['distance'].agg(['mean', 'std'])
stats
```

```python
pd.concat([new_df, stats], axis='columns')
```
