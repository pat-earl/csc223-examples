```python
import pandas as pd
```

```python
cars = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/DAAG/toycars.csv', index_col=0)
cars.head()
```

```python
cars['distance'] *= 1000
```

```python
stats = cars.groupby('angle')['distance'].agg(['mean', 'std'])
stats
```

```python
new_df = cars.pivot_table(index='angle', columns='car', values='distance')
```

```python
pd.concat([new_df, stats], axis=1)
```
